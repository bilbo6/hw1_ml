```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("salary_prediction.csv")

data_train = data["years"]
target_train = data["salary"]

data_train_bias = np.c_[np.ones([len(data_train), 1]), data_train]
```

```python
plt.scatter(data_train,target_train)
plt.show()
```

```python
print(data_train_bias.T.shape)
print(target_train)
target_train = target_train.values.reshape(len(target_train), 1)
```

```python
learning_rate = 0.01
epochs = 1000
m = len(data_train)

weights = np.random.randn(2,1)

for epoch in range(epochs):
    pred = data_train_bias.dot(weights)

    gradients = 2/m * data_train_bias.T.dot(data_train_bias.dot(weights) - target_train)
    weights = weights - learning_rate * gradients
    
    if epoch % 100 == 0:
        loss = np.mean((pred - target_train) ** 2)
        print(f"Epoch {epoch}: Loss = {loss:.4f}")
```

```python
weights
```

```python
pred = data_train_bias.dot(weights)

plt.scatter(data_train, target_train, color="blue", label="Actual Data")

plt.plot(data_train, pred, color="red", label="Regression Line")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression(gradient)")

plt.show()
```
