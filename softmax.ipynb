```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
print(iris.feature_names)
data=data[:, (2, 3)]
print(data)
```

```python
data_bias = np.c_[np.ones([len(data), 1]), data]
print(data_bias)
```

```python
np.random.seed(123456)

indices = np.random.permutation(len(data_bias))

train_size = int(0.6 * len(data_bias))
validation_size = int(0.2 * len(data_bias))
test_size = len(data_bias) - train_size - validation_size

train_indices = indices[:train_size]
validation_indices = indices[train_size:train_size + validation_size]
test_indices = indices[train_size + validation_size:]

data_train = data_bias[train_indices]
data_validation = data_bias[validation_indices]
data_test = data_bias[test_indices]

target_train = target[train_indices]
target_validation = target[validation_indices]
target_test = target[test_indices]

print(f"Train size: {len(data_train)}, Validation size: {len(data_validation)}, Test size: {len(data_test)}")
```

```python
target_train_onehot = np.zeros((len(target_train), 3))
target_train_onehot[np.arange(len(target_train)), target_train] = 1
target_train_onehot
```

```python
target_test_onehot = np.zeros((len(target_test), 3))
target_test_onehot[np.arange(len(target_test)), target_test] = 1
target_test_onehot
```

```python
target_validation_onehot = np.zeros((len(target_validation), 3))
target_validation_onehot[np.arange(len(target_validation)), target_validation] = 1
target_validation_onehot
```

```python
def softmax(x):
    exponent = np.exp(x)
    exponent_sums = np.sum(exponent, axis=1, keepdims=True)
    return exponent / exponent_sums
```

```python
weights = np.random.randn(3,3)
```

```python
learning_rate = 0.1
epochs = 10000
m = len(data_train)
best_loss = np.inf

for epoch in range(epochs):
    logits = data_train.dot(weights)
    target_prob = softmax(logits)
    error = target_prob - target_train_onehot
    gradients = 1/m * data_train.T.dot(error)
    weights = weights - learning_rate * gradients

    logits = data_validation.dot(weights)
    prob = softmax(logits)
    loss = -np.mean(np.sum(target_validation_onehot * np.log(prob), axis=1))
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")
    if loss < best_loss:
        best_loss = loss
    else:
        print(epoch - 1, best_loss)
        print(epoch, loss, "early stopping")
        break
```

```python
weights
```

```python
logits = data_validation.dot(weights)
prob = softmax(logits)
prediction = np.argmax(prob, axis=1)

accuracy_score = np.mean(prediction == target_validation)
print(accuracy_score)
```

```python
logits = data_test.dot(weights)
prob = softmax(logits)
prediction = np.argmax(prob, axis=1)

accuracy_score = np.mean(prediction == target_test)
print(accuracy_score)
```

```python
x0, x1 = np.meshgrid(
        np.linspace(0, 8, 500).reshape(-1, 1),
        np.linspace(0, 3.5, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]
X_new_with_bias = np.c_[np.ones([len(X_new), 1]), X_new]

logits = X_new_with_bias.dot(weights)
Y_proba = softmax(logits)
y_predict = np.argmax(Y_proba, axis=1)

zz1 = Y_proba[:, 1].reshape(x0.shape)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.plot(data[target==2, 0], data[target==2, 1], "g^", label="Iris virginica")
plt.plot(data[target==1, 0], data[target==1, 1], "bs", label="Iris versicolor")
plt.plot(data[target==0, 0], data[target==0, 1], "yo", label="Iris setosa")

from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])

plt.contourf(x0, x1, zz, cmap=custom_cmap)
contour = plt.contour(x0, x1, zz1, cmap=plt.cm.brg)
plt.clabel(contour, inline=1, fontsize=12)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 7, 0, 3.5])
plt.show()
```
